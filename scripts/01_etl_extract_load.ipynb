```python
import pandas as pd
import sqlite3
import glob
import os

folder_path = 'operating_hours_2025' 
database_name = 'toronto_transportation.db'
table_name = 'vehicle_operating_data'

# Connect to SQLite and remove the old sample data
print("Connecting to database and clearing old sample data...")
conn = sqlite3.connect(database_name)
cursor = conn.cursor()
cursor.execute(f"DROP TABLE IF EXISTS {table_name}") 

# Find all CSV files in the folder that start with "operating_hours_20" and sort them by month
csv_files = glob.glob('operating_hours_20*/*.csv')
csv_files.sort() 
print(f"Found {len(csv_files)} files. Starting the large data load... This may take a few minutes.\n")

# Loop through the files one by one to reduce memory usage
for file in csv_files:
    print(f"Loading {file}...")
    df = pd.read_csv(file)
    df.to_sql(table_name, conn, if_exists='append', index=False)
print("\nSuccess! All data from 2020 to 2026 has been loaded into SQLite.")

test_query = f"SELECT count(*) as total_rows FROM {table_name};"
count_df = pd.read_sql_query(test_query, conn)
print(f"\nTotal rows currently in the database: {count_df['total_rows'][0]:,}")

conn.close()
```
